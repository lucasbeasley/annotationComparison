```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# ***** BAR GRAPH OF TOTAL AND UNIQUE ANNOTATIONS WITHIN CORPUS *****
cat = ['BP', 'CC', 'MF']
total = [15747, 8310, 3508]
unique = [713, 204, 247]
x = range(1, 4)
ind = np.arange(len(x))
```

```python
fig = plt.figure(figsize=(12,9), dpi=1200)
ax = fig.add_subplot(111)
# set bars
ax.bar(ind+0.25, total, 0.45, color='#AFAFAF', align='center', label='Total')
ax.bar(ind, unique, 0.65, color='#000000', align='center', label='Unique')
# get the colors and labels for the legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, frameon=False, fontsize=24)
# set the x axis categories and label for y axis
plt.xticks(ind, cat, fontsize=24)
plt.yticks(fontsize=24)
plt.ylabel('Number of Annotations', fontsize=24)
# remove frame
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# save figure
plt.savefig(output + "GOdist.eps", format='eps', dpi=1200)
plt.show()
```
